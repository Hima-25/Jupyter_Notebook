```python
import matplotlib.pyplot as plt
import numpy as np
x= np.array (["a","b","c","d"])
y= np.array([400,200,900,100])
plt.bar(x,y , color ='green' , width = 0.5)
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.arange(-5,5,0.01)
y=x**2
plt.plot(x,y)

plt.annotate('function minimum \n relative to data' , xy=(0,0) , xycoords = 'data', xytext = (2,3), arrowprops = dict(facecolor ='black' , shrink =0.05) ,
             horizontalalignment ='left' ,verticalalignment='top')
plt.show()
```

```python
t=np.arange(0.0,5.0,0.001)
u=np.cos(3*np.pi*t)
plt.plot(t,u,lw=2)
plt.annotate('local maxima' , xy =(3,3.1),xytext =(3,1.8))
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.array (["a","b","c","d"])
y= np.array([400,200,900,100])
plt.barh(x,y , color ='green' , height = 0.5)  #horizontal
plt.show
```

```python
#Histogram
x = np.random.normal(170,10,250)
plt.hist(x)
plt.show()
```

```python
#Histogram
x = np.random.normal(170,10,250)
plt.hist(x ,bins=15 , color='red')
plt.show()
```

```python
#pichart
y=np.array([35,25,25,15])
plt.pie(y)
plt.show
```

```python
#pichart
y=np.array([35,25,25,15])
mylabels=['a','b','c','d']
mycolors = ['pink','blue','black','green']
myexplode= [0.2,0,0,0]
plt.pie(y , labels=mylabels ,startangle = 90 ,explode =myexplode , colors =mycolors , autopct = '%1.2f%%') #shadow = true)
plt.legend(title='four fruits',loc='upper right')
plt.show
```

```python
#3D PLOT
from mpl_toolkits import mplot3d
%matplotlib notebook
import numpy as mp
import matplotlib.pyplot as plt
fig = plt.figure
ax= plt.axes(projection = '3d')
z= np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax.plot3D(x,y,z,'red' ,linewidth =5)
ax.set_title('3D plot')
plt.show()
```

```python
from mpl_toolkits import mplot3d
%matplotlib notebook
import numpy as mp
import matplotlib.pyplot as plt
#creating dataset
fig = plt.figure(figsize =(10,5))
ax = plt.axes(projection ='3d')

z=np.random.randint(100,size=(100))
x=np.random.randint(150,size=(100))
y=np.random.randint(800,size=(100))

colors=np.random.randint(100,size = (100))
#creating 3d plot
ax.scatter3D(x,y,z , c=colors , alpha = 0.9 , cmap='rainbow' ,marker = 's')
ax.set_title ('Scatter_3D')
plt.show()
```
