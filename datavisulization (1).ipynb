# LINE

family

[ 'serif' | 'sans-serif' | 'cursive' | 'fantasy' | 'monospace' ]

```python
x=[5,10,15,20,25,30,35,40,45]
y=[2,5,7,1,3,8,4,6,8]
plt.plot(x,y,linestyle = '--' , color ='red' , linewidth = '2.1')  
font1= {'family': 'serif', 'color':'blue','size' :20 }
font2= {'family': 'serif', 'color':'darkred','size' :20}
plt.xlabel('x-label', fontdict = font1)
plt.ylabel('y-label', fontdict = font2)

plt.tick_params(axis='x',labelcolor = 'green' ,labelsize = '18' , labelrotation = -60 , length =14.2)
plt.tick_params(axis='y',labelcolor = 'red' ,labelsize = '18' , labelrotation = 30 , length = 14.2)


plt.title("trail plot" ,fontdict = font1 ,loc= 'center')
plt.grid(axis = 'x', color = 'green' , linestyle = '--' , linewidth =1.5)
plt.show
```

# LEGEND

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x), '--b', label = 'sineplot')
plt.plot(x,np.cos(x), c='r' , label = 'cosine')
plt.legend(loc = 'upper right' , bbox_to_anchor = (0.75,1.15) , shadow = bool)
plt.show
```

```python
#OR

x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x), '--b')
plt.plot(x,np.cos(x), c='r' )
plt.legend(['sine','cosine'],loc = 'upper right' , bbox_to_anchor = (0.75,1.15) , shadow = bool)
plt.show
```

# 3D PLOT


```python
from mpl_toolkits import mplot3d
%matplotlib notebook
fig = plt.figure
ax= plt.axes(projection = '3d')
z= np.linspace(0,1,100)
x=z*np.sin(25*z)
y=z*np.cos(25*z)
ax.plot3D(x,y,z,'red' ,linewidth =5)
ax.set_title('3D plot')
plt.show()
```

```python
#creating dataset
fig = plt.figure(figsize =(10,5))
ax = plt.axes(projection ='3d')

z=np.random.randint(100,size=(100))
x=np.random.randint(150,size=(100))
y=np.random.randint(800,size=(100))

colors=np.random.randint(100,size = (100))
#creating 3d plot
ax.scatter3D(x,y,z , c=colors , alpha = 0.9 , cmap='rainbow' ,marker = 's')
ax.set_title ('Scatter_3D')
plt.show()
```
